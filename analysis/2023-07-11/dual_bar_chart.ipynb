```python
import matplotlib.pyplot as plt
import numpy as np

%config InlineBackend.figure_format = 'retina'  #  jupyter notebook 
# plt.style.use('ggplot')

loss_proportion = [32.934089, 5.354427, 5.037901, 4.624707, 3.858441, 3.162908, 2.696076, 2.335341, 2.233959, 1.410518]
avg_plr = [7.05286, 4.60075, 0.49934, 1.15891, 0.0203, 0.10362, 0.05891, 0.37489, 1.11909, 0.18057]

# X
x = np.arange(1, 11)

# figure
fig, ax1 = plt.subplots()

# 
ax1.bar(x, loss_proportion, color='tab:blue', alpha=0.7)
ax1.set_xticks(x, x)

# y
ax1.set_ylabel('Loss Proportion (%)', color='tab:blue')
ax1.set_title('Downlink Loss Proportion and Average PLR')

# x
ax2 = ax1.twinx()

# 
ax2.plot(x, avg_plr, color='tab:red', marker='o')

# y
ax2.set_ylabel('Average PLR (%)', color='tab:red')

# x
plt.xlabel('Item')

# 
plt.show()

print(sum(loss_proportion))
```
