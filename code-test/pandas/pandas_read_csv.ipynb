```python
import pandas as pd
```

```python
df = pd.read_csv("data/diag_log_qc03_2023-02-04_16-50-15_ml1.csv")
display(df.info())
display(df.head())
```

```python
df = pd.read_csv("data/diag_log_qc03_2023-02-04_16-50-15_ml1.csv",
                 dtype={'type_id':'category',
                        'Serving Cell Index':'category'},
                 parse_dates=['Timestamp'])
display(df.info())
display(df.head())
```

```python
dtypes = {'type_id':'category', 'Serving Cell Index':'category'}
parse_dates = ['Timestamp']
df = pd.read_csv("data/diag_log_qc03_2023-02-04_16-50-15_ml1.csv", dtype=dtypes, parse_dates=parse_dates)
display(df.info())
display(df.head())
```
